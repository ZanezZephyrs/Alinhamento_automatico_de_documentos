```python
path="data_tiffs"
dest="imagens_src"
for root, dirs, files in os.walk(path):
    for file in files:
        if ".tif" in file:
            print(root)
            print("separando imagem {}".format(file))
            img = Image.open(root+"/"+file)
            for i in range(10):
                try:
                    img.seek(i)
                    img.save(dest+"/"+file[:-3]+'page_%s.tif'%(i,))
                except:
                    break
```
